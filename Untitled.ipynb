```python
RFModel = RandomForestClassifier(random_state=0, n_estimators=250, class_weight={0:.96, 1:.04})
RFModel.fit(X_train, y_train)

# RFModel = svm.SVC()
# RFModel.fit(X_train, y_train)

GBModel = GradientBoostingClassifier(random_state=0, n_estimators=250)
GBModel.fit(X_train, y_train)
```
